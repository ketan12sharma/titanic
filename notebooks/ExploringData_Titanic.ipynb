```python
# Centrality Measure i.e mean and median
'''
Represented by a single number for entire values
The number will be central to data
'''
df.Fare.mean() 
# Mean = Sum of All Values / Number of Values 
# Problems: 
# 1. Gets affected by extreme values i.e. mean(6,7,8) = 7 But if we add an extreme value mean(6,7,8,51) 
# than mean becomes 18 
```

```python
df.Fare.median()
# Solves the extreme value problem by sorting the list and taking the middle value 
# Median(6,7,8) = 7 and (6,7,8,51) = 7.5
```

```python
# Dispersion Measure 
'''
Describes how spread out the values are from a central value
Gives the variablity of data that how similar/ dissimilar values are there
'''
df.Fare.max() - df.Fare.min()

# Range of the Fare
# Range is directly proportional to data spread
# Range is affected by extreme values i.i. range(6, 7, 8) = 2 
# range(6,7,8 ,51 ) =  45 which is incorrect as 75% values are less than 10.
```

```python
# Percentiles: x percentile y means that x% values are less than y.
# The percentiles 25, 50 and 75 are described as quartiles
print('25th Percentile: {0}'.format(df.Fare.quantile(.25))) #25th Percentile 
print('50th Percentile: {0}'.format(df.Fare.quantile(.50)))
print('75th Percentile: {0}'.format(df.Fare.quantile(.75)))
```

```python
# Variance: Measure of variablity
# How far each value in list from mean value
# Variance is directly proportional to spread
# Variance = sum((value - mean)^2)/count
# Affected by extreme values and the units are also not clear
print('Variance: {0}'.format(df.Fare.var()))
# to overcome the problems with variance we use standard deviation
# Std. Dev = sqrt(variance) 
print('Standard Deviation: {0}'.format(df.Fare.std()))
```
