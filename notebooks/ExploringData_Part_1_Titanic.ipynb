```python
# Centrality Measure i.e mean and median
'''
Represented by a single number for entire values
The number will be central to data
'''
df.Fare.mean() 
# Mean = Sum of All Values / Number of Values 
# Problems: 
# 1. Gets affected by extreme values i.e. mean(6,7,8) = 7 But if we add an extreme value mean(6,7,8,51) 
# than mean becomes 18 
```

```python
df.Fare.median()
# Solves the extreme value problem by sorting the list and taking the middle value 
# Median(6,7,8) = 7 and (6,7,8,51) = 7.5
```

```python
# Dispersion Measure 
'''
Describes how spread out the values are from a central value
Gives the variablity of data that how similar/ dissimilar values are there
'''
df.Fare.max() - df.Fare.min()

# Range of the Fare
# Range is directly proportional to data spread
# Range is affected by extreme values i.i. range(6, 7, 8) = 2 
# range(6,7,8 ,51 ) =  45 which is incorrect as 75% values are less than 10.
```

```python
# Percentiles: x percentile y means that x% values are less than y.
# The percentiles 25, 50 and 75 are described as quartiles
print('25th Percentile: {0}'.format(df.Fare.quantile(.25))) #25th Percentile 
print('50th Percentile: {0}'.format(df.Fare.quantile(.50)))
print('75th Percentile: {0}'.format(df.Fare.quantile(.75)))
```

```python
# Variance: Measure of variablity
# How far each value in list from mean value
# Variance is directly proportional to spread
# Variance = sum((value - mean)^2)/count
# Affected by extreme values and the units are also not clear
print('Variance: {0}'.format(df.Fare.var()))
# to overcome the problems with variance we use standard deviation
# Std. Dev = sqrt(variance) 
print('Standard Deviation: {0}'.format(df.Fare.std()))
```

## Using Group By


```python
# group by single column and apply the summary functionr
# 1. Group by returns group similar values for a particular column
# 2. Select the column you want to aggregate the results by e.g. here Age
# 3. Apply the aggregation function

df.groupby('Sex').Age.median()
```

```python
df.groupby(['Pclass']).Fare.median()
```

```python
df.groupby(['Pclass']).Age.median()
```

```python
# Multiple summary aggregation on the same groupby columns
df.groupby(['Pclass'])['Fare','Age'].median()
```

```python
# Aggregations for more than one type
df.groupby(['Pclass']).agg({'Fare': 'mean', 'Age':'median'})
```

```python
# More complex aggregation:
aggregations = {
    'Fare' : { # Select Fare Column
        'mean_fare': 'mean',
        'median_fare': 'median',
        'max_fare': max,
        'min_fare': np.min
    },
    'Age' : { # Select Age Column
        'median_age': 'median',
        'min_age': min,
        'max_age': max,
        'range_age': lambda x: max(x) - min(x) # calculate range per group
    },
}
```

```python
df.groupby(['Pclass']).agg(aggregations)
```

```python
df.groupby(['Pclass', 'Embarked']).Fare.median()
```

## Pivot Table

```python
'''Informationn required to make the Pivot Table:
    1. Selecting the attribute/ column for pivot row (Sex) index
    2. Selecting the attribute/ column for pivot column (Pclass) columns 
    3. Selecting the attribute column for filling the cells (Age) values
    4. Selecting the attribute column for applying the aggregate function (Mean / average) aggfunc'''
df.pivot_table(index = 'Sex', columns = 'Pclass', values = 'Age', aggfunc = 'mean')
```

```python
# pivot table using group by
df.groupby(['Sex', 'Pclass']).Age.mean()
```

```python
df.groupby(['Sex', 'Pclass']).Age.mean().unstack()
```
